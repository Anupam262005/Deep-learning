```python
# Import required libraries from TensorFlow / Keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, AveragePooling2D, Flatten, Dense

# ---------------------------------------------------------
# Create a Sequential model
# Sequential means layers will be added one after another
# ---------------------------------------------------------
model = Sequential()

# ---------------------------------------------------------
# C1: First Convolution Layer
# filters = 6  → number of feature maps (kernels)
# kernel_size = (5,5) → size of convolution filter
# activation = 'tanh' → activation function used in original LeNet
# padding = 'valid' → no padding (output shrinks)
# input_shape = (32,32,1) → 32x32 grayscale image
# ---------------------------------------------------------
model.add(Conv2D(filters=6,
                 kernel_size=(5,5),
                 activation='tanh',
                 padding='valid',
                 input_shape=(32,32,1)))

# ---------------------------------------------------------
# S2: Average Pooling Layer
# pool_size = (2,2) → pooling window size
# strides = 2 → window moves 2 pixels each step
# padding = 'valid' → no padding
# Purpose: reduce spatial size and computation
# ---------------------------------------------------------
model.add(AveragePooling2D(pool_size=(2,2),
                           strides=2,
                           padding='valid'))

# ---------------------------------------------------------
# C3: Second Convolution Layer
# filters = 16 → 16 feature maps
# kernel_size = (5,5)
# activation = 'tanh'
# padding = 'valid'
# ---------------------------------------------------------
model.add(Conv2D(filters=16,
                 kernel_size=(5,5),
                 activation='tanh',
                 padding='valid'))

# ---------------------------------------------------------
# S4: Second Average Pooling Layer
# Again reduces spatial dimension by half
# ---------------------------------------------------------
model.add(AveragePooling2D(pool_size=(2,2),
                           strides=2,
                           padding='valid'))

# ---------------------------------------------------------
# Flatten Layer
# Converts 3D feature maps into a 1D vector
# Required before feeding into Dense (fully connected) layers
# ---------------------------------------------------------
model.add(Flatten())

# ---------------------------------------------------------
# Fully Connected Layer (FC1)
# 120 neurons
# activation = tanh
# ---------------------------------------------------------
model.add(Dense(120, activation='tanh'))

# ---------------------------------------------------------
# Fully Connected Layer (FC2)
# 84 neurons
# activation = tanh
# ---------------------------------------------------------
model.add(Dense(84, activation='tanh'))

# ---------------------------------------------------------
# Output Layer
# 10 neurons → for 10 classes (digits 0-9 in MNIST)
# softmax converts outputs into probabilities
# ---------------------------------------------------------
model.add(Dense(10, activation='softmax'))

# ---------------------------------------------------------
# Compile the model
# optimizer = adam → adjusts weights during training
# loss = categorical_crossentropy → used for multi-class classification
# metrics = accuracy → performance measure
# ---------------------------------------------------------
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# ---------------------------------------------------------
# Print the model architecture summary
# Shows layer shapes and number of parameters
# ---------------------------------------------------------
model.summary()
```
